```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 此行代码，防止绘图警告信息
%matplotlib inline 
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python


fig, a = plt.subplots() # 创建一个子图
plt.figure(figsize=(16,9))
a.plot(x, y, label='linear')
# a.plot(x, x**2, label='quadratic')

ax.text(.25, .025, r'$\mu=115,\ \sigma=15$')

a.set_xticks([0,1,2,3,4,5,6]) # 先设置索引横坐标刻度
a.set_xticklabels(['zero','one','two','three','four','five','six'], rotation = 30)

a.set_title('Simple plot')
a.set_xlabel('time')
a.set_ylabel('data')
# plt.ylabel('data')
plt.tight_layout() # 调整图表
a.legend();
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
plt.tight_layout();
```
